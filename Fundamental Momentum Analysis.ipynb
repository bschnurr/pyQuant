```python
import price
import pandas as pd
from fmp_db import fmp
pd.options.display.float_format = '{:.4f}'.format
```

```python
db = fmp()
```

```python
df = db.load_financials()
df
```

Company Size

```python
df['marketCap'] = df['marketCap']/100000000
df
```

```python
minMktCap = 1000
```

```python
df_mktcap = df[df['marketCap'] > minMktCap]
df_mktcap
```

Performance Matrices

```python
minRevGrowth = 0.3
minGPMargin = 0.2
minEPSGrowth = 0.1
minROE = 0.2
```

```python
conditions = (df_mktcap['revenue_per_share_fiveY_growth'] > minRevGrowth) & (df_mktcap['gross_profit_margin'] > minDPSGrowth) & (df_mktcap['eps_growth'] > minEPSGrowth) & (df_mktcap['roe'] > minROE)
```

```python
df_performance = df_mktcap[conditions]
```

```python
df_performance
```

Dividend Matrices

```python
minDPSGrowth = 0.1
minDivYield = 0
```

```python
div_conds = (df_performance['dps_fiveY_growth'] > minDPSGrowth) & (df_performance['div_yield'] > minDivYield)
```

```python
df_final = df_performance[div_conds]
df_final
```

```python
mom_data = {
    'symbol': [],
    'currentPrice': [],
    'recentHigh': [],
    '15%_discount': [],
    '20%_discount': [],
    '25%_discount': []
}
```

```python
for symbol in list(df_final['symbol']):
    print(symbol)
    currPrice = db.get_current_price(symbol)
    high = price.calculate_prev_max_high(symbol, 90)
    fifteen = high * 0.85
    twenty = high * 0.80
    twentyFive = high * 0.75
    
    mom_data['symbol'].append(symbol)
    mom_data['currentPrice'].append(currPrice)
    mom_data['recentHigh'].append(high)
    mom_data['15%_discount'].append(fifteen)
    mom_data['20%_discount'].append(twenty)
    mom_data['25%_discount'].append(twentyFive)
```

```python
mom_df = pd.DataFrame(mom_data)
mom_df
```
